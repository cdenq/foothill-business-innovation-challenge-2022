```python
import matplotlib.pyplot as plt
```

```python
a = [0.95, .75, .67, .24]
b = ['Goal', 'Gen 3', 'Gen 2','Gen 1']
```

```python
plt.bar(b,a)
plt.title("Prediction Strength of FAIR SCORE Algorithm")
plt.ylabel("Prediction Strength")
plt.xlabel("Algorithm Attempts")
plt.grid(alpha=0.4)
for i in range(len(a)):
    plt.text(i-.15,a[i]+0.01,a[i])
```
